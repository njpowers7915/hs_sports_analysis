# High School Sports Analysis

```python
import pandas as pd
participation_df = pd.read_excel('hs_sports_data.xlsx', index_col = 0)
```

```python
participation_df.head()
```

```python
sports = sorted(list(participation_df['Sport'].value_counts().keys()))
sports
```

```python
analyzed_sports = ['Baseball', 'Basketball', 'Ice Hockey', 'Football -- 11-Player', 'Football -- 6-player', 'Football -- 8-player', 'Football -- 9-player', 'Soccer', 'Tennis', 'Track and Field -- Indoor', 'Track and Field -- Outdoor']
participation_df = participation_df[participation_df['Sport'].isin(analyzed_sports)]
participation_df.head(11)
```

```python
participation_df = participation_df.drop(columns=['Boys School', 'Girls School', 'Girls Participation'])
participation_df = participation_df.rename(columns={"Boys Participation": "Participation"})
participation_df = participation_df.reset_index()
participation_df.head()
```

```python
school_year_start = participation_df['Year'].str.split("/", n = 1, expand = True)
participation_df['Year'] = school_year_start[0]

football = participation_df['Sport'].str.split(" -- ", n = 1, expand = True)
participation_df['Sport'] = football[0]
participation_df.head()
#participation_df = participation_df.set_index('Sport')
#participation_df.groupby(['Sport', 'State', 'Year']).sum()

#new = data["Name"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
#data["First Name"]= new[0] 
```

```python
#participation_df = participation_df.groupby(['Sport', 'State'])
participation_df.head()
sport_df = participation_df[participation_df['Sport'] == 'Baseball']
sport_df.head()
state_df = sport_df[sport_df['State'] == 'AL']
state_df.pivot(index='State', columns='Year', values='Boys Participation')

sports = {}
def measure_growth_by_state(sport, state):
    sport_df = participation_df[participation_df['Sport'] == sport]
    state_df = sport_df[sport_df['State'] == state]
    return state_df.pivot(index='State', columns='Year', values='Boys Participation')

measure_growth_by_state('Basketball', 'MA')
```
